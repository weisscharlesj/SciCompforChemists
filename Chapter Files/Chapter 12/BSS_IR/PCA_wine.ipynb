# Unsupervised Dimensionality Reduction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
wine = load_wine()
```

```python
wine.data
```

```python
wine.target
```

```python
pca = PCA(n_components=2)
pca.fit(wine.data)
components = pca.transform(wine.data)
```

```python
components.shape
```

```python
plt.scatter(components[:,0], components[:,1], c=wine.target)
```

## t-SNE

```python
tsne = TSNE(n_components=2)
tcomp = tsne.fit_transform(wine.data)
```

```python
tcomp.shape
```

```python
plt.scatter(tcomp[:,0], tcomp[:,1], c=wine.target)
```
